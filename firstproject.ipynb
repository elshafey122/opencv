# imports



```python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import numpy as np
```

#*load* image

```python
image=cv2.imread('/content/istockphoto-849395194-612x612.jpg')
image= cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
cv2_imshow(image)
```

# describe image and do changes on it

```python

print(type(image.shape))
print("shape",image.shape);
print("image.dtype",image.dtype)
print("image.min",image.min())
print("image.max",image.max())
print("image.size",image.size)


```

**crop image with indicate part of image**


```python
image[1:10,1:10,]
cropper_image= image[:,:800,]
cv2_imshow(cropper_image)
```

**resizing the image**

```python
resize_image= cv2.resize(image,(300,300))
cv2_imshow(resize_image)
```

**change color to gray image**

```python
gray_img= cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img);
```
